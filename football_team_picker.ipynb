```python
import pandas as pd
import tensorflow as tf
```

```python
filenames = ['./season_25.csv','./season_24.csv','./season_20.csv', './season_19.csv', './season_18.csv']

csv = pd.read_csv(filenames[0])
csv
```

```python
total_player_list = []
flipped_dfs = []

for filename in filenames:
    csv = pd.read_csv(filename)

    # remove rows and columns we don't need
    cleaned_csv = csv.dropna(subset=['name'])
    cleaned_csv = cleaned_csv.drop(labels=['wins', 'apps', 'win ratio'], axis=1)

    #flip it to be the way we want
    flipped = cleaned_csv.set_index('name').transpose()

    #drop date column for now (has been renamed 'name' in the flip)
    # I might use this in future work, but not now
    flipped_dfs.append(flipped)
    total_player_list = total_player_list + list(flipped.columns)

total_player_list = list(set(total_player_list)) # make to a unique list

print(total_player_list)

## Add to each frame, so that they all have the same columns
for fd in flipped_dfs:
    for name in total_player_list:
        if name not in list(fd.columns):
            fd.insert(len(fd.columns), name, [float("NaN")] * len(fd), True) 
    
flipped_dfs

flipped = pd.concat(flipped_dfs)
flipped
```

```python
player_names = list(flipped.columns) 
player_names
```

```python
RANDOMER = "Randomer"

def formRandomerName(num: int) -> str:
    return RANDOMER + str(num)
    

## work out how many randomers we need

totals = []
for index, row in flipped.iterrows():
    losers = list(row.array).count(0)
    winners = list(row.array).count(1)
    drawers =  list(row.array).count(0.5)
    total_players = losers + winners + drawers
    totals.append(total_players)

print("average players per game: %f" % (sum(totals)/len(totals)))
min_players = min(totals)
randomers_needed = 10-min_players
print("Minimum numbers of players present in a game: %i" % min_players)
print("Maximum number of randomers in a game: %i" % randomers_needed)

for i in range(randomers_needed):
    randomer_name =  formRandomerName(i)
    flipped[randomer_name] = float("NaN")


# Now populate results for these randomers
for index, row in flipped.iterrows():
    losers = list(row.array).count(0)
    winners = list(row.array).count(1)
    drawers =  list(row.array).count(0.5)
    total_players = losers + winners + drawers
    totals.append(total_players)
    
    randomer_index = 0
    if drawers != 0:
        for i in range(10-drawers):
            randomer_name =  formRandomerName(randomer_index)
            row[randomer_name] = 0.5
            randomer_index+=1
        continue #If its a draw, there should be no winners or losers

    for i in range(5-winners):
        randomer_name =  formRandomerName(randomer_index)
        row[randomer_name] = 1
        randomer_index+=1
        
    for i in range(5-losers):
        randomer_name =  formRandomerName(randomer_index)
        row[randomer_name] = 0
        randomer_index+=1
        
# Check that everything sums up ok, print if it doesn't
for index, row in flipped.iterrows():
    losers = list(row.array).count(0)
    winners = list(row.array).count(1)
    drawers =  list(row.array).count(0.5)
    total_players = losers + winners + drawers
    if total_players != 10:
        print(row)

player_names = list(flipped.columns) 
        
flipped
```

```python
import math
import random

def is_picked(value: float, winning_team: bool) -> float:
    if math.isnan(value):
        return 0.0
    if value == 0.5:
        #draw game, we will assign these later
        return value
    if winning_team:
        return value
    else:
        return 1-value
            

# we call the winners the white team, and the losers the black team
for player in player_names:
    flipped[player+'_white'] = flipped[player].apply(lambda x:is_picked(x, True))
    flipped[player+'_black'] = flipped[player].apply(lambda x:is_picked(x, False))

#show how we decomposed this
flipped[['Paul', 'Paul_white', 'Paul_black']]
```

```python
def get_result(row):
    if 0.5 in list(row):
        return 0.5
    else:
        return 1.0


#add result. These are all white wins, unless they are draws
flipped['result'] = flipped.apply(get_result , axis=1)


flipped[['Paul', 'Paul_white', 'Paul_black', 'result']]
```

```python
#fill in draws, randomly picking teams, as we don't have this info
import random

for index in flipped.index:
    if flipped.loc[index,'result'] == 0.5:
        white_players = 0
        #pick 5 guys who were there for white team
        while white_players < 5:
            player = random.choice(player_names)
            if flipped.loc[index, player+'_white'] == 0.5:
                flipped.loc[index, player+'_white'] = 1.0
                flipped.loc[index, player+'_black'] = 0.0
                white_players += 1
        #put all the others playing into the black team
        for player in player_names:
            if flipped.loc[index, player+'_black'] == 0.5:
                flipped.loc[index, player+'_white'] = 0.0
                flipped.loc[index, player+'_black'] = 0.1
                white_players += 1
            
print(flipped[['Paul', 'Paul_white', 'Paul_black', 'result']])

flipped

    
```

```python
## WEATHER

from src.weather.weather_service import WeatherService
from datetime import datetime, timedelta

FORMAT = '%d/%m/%Y %H:%M'

BELFIELD_LATITUDE = 53.3067
BELFIELD_LONGITUDE = -6.2210


weather_service = WeatherService(BELFIELD_LATITUDE, BELFIELD_LONGITUDE)
weather_service.set_station("03969")

def formStartEndTimes(datestr: str):
    if datestr[2] != '/':
        datestr = '0' + datestr
    if datestr[5] != '/':
        datestr = datestr[0:3] +'0' + datestr[3:]
    
    datestr = datestr + ' 21:00' # our matches have always started at 9pm
    start = datetime.strptime(datestr, FORMAT)

    end = start + timedelta(hours=1)

    return start, end
    

def getRainfall(row):
    datestr = row.name
    start, end = formStartEndTimes(datestr)
    weather = weather_service.get_weather(start, end)
    return weather.rain

flipped['rainfall'] = flipped.apply(getRainfall, axis=1)
```

```python
#drop originals
white_wins = flipped.drop(player_names, axis=1)
```

```python
#make black wins dataset, which are the same results, but with the labels flipped, and black always winning
# this is important to make sure the algo doesn't always think that the white team wins every time!!!

mapping = {}
for player in player_names:
    mapping[player+ '_white'] = player + '_black'
    mapping[player+ '_black'] = player + '_white'
        

black_wins = white_wins.rename(columns=mapping)
black_wins['result'] = black_wins['result'].apply(lambda x:1-x)

black_wins
```

```python
combined = pd.concat([white_wins, black_wins])

#index has  dupes after the combine, need to remove these
combined.index = range(len(combined))
combined
```

```python
## Verify that there are no Nans in the frame, as this will cause serious issues

for col in combined.columns:
    print(col +
    " type: " + str(combined.dtypes[col]) +
    " Nan count: " + str(combined[col].isna().sum()))
```

```python
target = combined.pop('result')
target
```

```python
dataset = tf.data.Dataset.from_tensor_slices((combined.values, target.values))

for feat, targ in dataset.take(5):
  print ('Features: {}, Target: {}'.format(feat, targ))

dataset = dataset.shuffle(len(combined)).batch(1)

## Need to maximise our training dataset, so we just use a subset for validation
train_size = int(0.9 * len(dataset))

train_dataset = dataset
val_dataset = dataset.skip(train_size)

print("Total rows: " + str(len(dataset)))
print("Training set: " + str(len(train_dataset)))
print("Validation set: " + str(len(val_dataset)))
```

```python
tf.keras.backend.set_floatx('float64')

def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError() ,
                metrics=['accuracy'])
  return model

model = get_compiled_model()
model.fit(train_dataset,validation_data=val_dataset, epochs=12)
```
