```python
import pandas as pd
import tensorflow as tf
```

```python
filenames = ['./season_20.csv', './season_19.csv', './season_18.csv']

csv = pd.read_csv(filenames[0])
csv
```

```python
total_player_list = []
flipped_dfs = []

for filename in filenames:
    csv = pd.read_csv(filename)

    # remove rows and columns we don't need
    cleaned_csv = csv.dropna(subset=['name'])
    cleaned_csv = cleaned_csv.drop(['wins', 'apps', 'win ratio'], 1)

    #flip it to be the way we want
    flipped = cleaned_csv.set_index('name').transpose()

    #drop date column for now (has been renamed 'name' in the flip)
    # I might use this in future work, but not now
    flipped_dfs.append(flipped)
    total_player_list = total_player_list + list(flipped.columns)

total_player_list = list(set(total_player_list)) # make to a unique list

print(total_player_list)

for fd in flipped_dfs:
    for name in total_player_list:
        if name not in list(fd.columns):
            fd.insert(len(fd.columns), name, [float("NaN")] * len(fd), True) 
    
flipped_dfs

flipped = pd.concat(flipped_dfs)
flipped
```

```python
player_names = list(flipped.columns) 
player_names
```

```python
## TODO: add randomers here
```

```python
import math
import random

def is_picked(value: float, winning_team: bool) -> float:
    if math.isnan(value):
        return 0.0
    if value == 0.5:
        #draw game, we don't have any team info, just pick randomly who was on what teams
        return random.choice([0.0, 1.0])
    if winning_team:
        return value
    else:
        return 1-value
            

# we call the winners the white team, and the losers the black team
for player in player_names:
    flipped[player+'_white'] = flipped[player].apply(lambda x:is_picked(x, True))
    flipped[player+'_black'] = flipped[player].apply(lambda x:is_picked(x, False))

#show how we decomposed this
flipped[['Paul', 'Paul_white', 'Paul_black']]
```

```python
def get_result(row):
    if 0.5 in list(row):
        return 0.5
    else:
        return 1.0


#add result. These are all white wins, unless they are draws
flipped['result'] = flipped.apply(get_result , axis=1)

flipped[['Paul', 'Paul_white', 'Paul_black', 'result']]
```

```python
#drop originals
white_wins = flipped.drop(player_names, 1)
```

```python
#make black wins dataset, which are the same results, but with the labels flipped, and black always winning
# this is important to make sure the algo doesn't always think that the white team wins every time!!!

mapping = {}
for player in player_names:
    mapping[player+ '_white'] = player + '_black'
    mapping[player+ '_black'] = player + '_white'
        

black_wins = white_wins.rename(columns=mapping)
black_wins['result'] = black_wins['result'].apply(lambda x:1-x)

black_wins
```

```python
combined = pd.concat([white_wins, black_wins])

#index has  dupes after the combine, need to remove these
combined.index = range(len(combined))
combined
```

```python
for col in list(combined.columns):
    print(col +
    " type: " + str(combined.dtypes[col]) +
    " Nan count: " + str(combined[col].isna().sum()))
```

```python
target = combined.pop('result')
target
```

```python
dataset = tf.data.Dataset.from_tensor_slices((combined.values, target.values))
#dataset = dataset.shuffle(len(df)).batch(1)

for feat, targ in dataset.take(5):
  print ('Features: {}, Target: {}'.format(feat, targ))

dataset = dataset.shuffle(len(combined)).batch(1)

train_size = int(0.7 * len(dataset))



train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)

#print("Total rows: " + str(len(dataset)))
#print("Training set: " + str(len(train_dataset)))
#print("Validation set: " + str(len(val_dataset)))
```

```python
tf.keras.backend.set_floatx('float64')

def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(15, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
  ])

  model.compile(optimizer='adam',
                loss=tf.keras.losses.MeanSquaredError() ,
                metrics=['accuracy'])
  return model

model = get_compiled_model()
model.fit(train_dataset,validation_data=val_dataset, epochs=12)
```
